```python
# Menentukan nama file CSV yang akan digunakan untuk menyimpan hasil pencarian tweet
filename = 'bahlil.csv'

# Menentukan kata kunci pencarian tweet
# Format pencarian: "bahlil since:2025-01-01 until:2025-03-07 lang:id"
# - "bahlil": Kata kunci yang dicari
# - "since:2025-01-01": Mencari tweet mulai dari tanggal 1 Januari 2025
# - "until:2025-03-07": Mencari tweet hingga tanggal 7 Maret 2025
# - "lang:id": Hanya mencari tweet dalam bahasa Indonesia
search_keyword = 'bahlil since:2025-01-01 until:2025-03-07 lang:id'

# Menentukan batas jumlah tweet yang akan diambil (dalam hal ini, 100 tweet)
limit = 100

# Menjalankan perintah untuk mengumpulkan tweet menggunakan tool tweet-harvest
# - `npx`: Menjalankan paket Node.js langsung tanpa perlu menginstalnya secara global
# - `-y`: Secara otomatis menyetujui semua prompt (yes)
# - `tweet-harvest@2.6.1`: Menggunakan versi 2.6.1 dari tool tweet-harvest
# - `-o "{filename}"`: Menyimpan hasil pencarian ke dalam file dengan nama yang ditentukan (bahlil.csv)
# - `-s "{search_keyword}"`: Menggunakan kata kunci pencarian yang telah ditentukan
# - `--tab "LATEST"`: Mengambil tweet berdasarkan tab "LATEST" (tweet terbaru)
# - `-l {limit}`: Membatasi jumlah tweet yang diambil sesuai dengan nilai `limit`
# - `--token {twitter_auth_token}`: Menggunakan token autentikasi Twitter untuk mengakses API
!npx -y tweet-harvest@2.6.1 -o "{filename}" -s "{search_keyword}" --tab "LATEST" -l {limit} --token {twitter_auth_token}
```

```python
import pandas as pd  # Mengimpor library pandas untuk manipulasi data

# Menentukan path file CSV yang berisi data tweet
file_path = f"tweets-data/{filename}"

# Membaca file CSV ke dalam dataframe 'df' dengan pemisah koma
df = pd.read_csv(file_path, delimiter=",")

# Menampilkan isi dataframe 'df' untuk melihat data yang telah dimuat
display(df)
```
